```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from datetime import datetime
import numpy as np
```

```python
data = pd.read_csv('tesla.csv')
data.head()
data.tail()
```

```python
# Convert the 'ID' column to datetime format, coercing invalid entries to NaT (Not a Time)
data['date'] = pd.to_datetime(data['ID'], errors='coerce')

# Extract Year, Month, and Day from the 'date' column (only if the conversion was successful)
data['Year'] = data['date'].dt.year
data['Month'] = data['date'].dt.month
data['Day'] = data['date'].dt.day

```

```python
#data['DateNumeric'] = data['Date'].apply(lambda x: x.toordinal())


```

```python
x = data[['Year', 'Month', 'Day']]
y = data['Close']
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values in the features (X)
imputer = SimpleImputer(strategy='mean')
x_imputed = imputer.fit_transform(x)

y = pd.to_numeric(y, errors='coerce')
y = y.dropna()  # Drop rows with NaN values in y


# Impute missing values in the target variable (y)

y_imputed = SimpleImputer(strategy='mean').fit_transform(y.values.reshape(-1, 1))



# Ensure that X is synchronized with y (remove corresponding rows in X if y has NaN)
x = x.loc[y.index]
```

```python

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
print(predictions)
```

```python
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(y_test, predictions)
print(rmse)
```

```python


# Transform the date into a single feature (ordinal representation)
date_to_predict = datetime(2025, 1, 1).toordinal() # Convert to ordinal

predict_date = datetime(2025, 1, 1)
predict_year = predict_date.year
predict_month = predict_date.month
predict_day = predict_date.day

# Reshape the date feature into the same format used during training
date_to_predict = np.array([[predict_year, predict_month, predict_day]])
print(date_to_predict)

# Make the prediction
new_pred = model.predict(date_to_predict)

# Output prediction
print(f"Predicted Stock on 2025, 1, 1: {new_pred[0]}")
print("The predicted value is 179.48 while actual value is around 400. This implies that Linear Regression is not suitable for stock prediction")
```
