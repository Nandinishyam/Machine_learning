```python
import pandas as pd #for file handing
import numpy as np #numerical resources
import matplotlib.pyplot as plt #visualisation
import seaborn as sns #visualisation
from sklearn.datasets import load_iris #ready made datasets
from sklearn.model_selection import train_test_split #alloting data for splitting training and testing
from sklearn.preprocessing import StandardScaler # rescaling the data between the given range
from sklearn.linear_model import LogisticRegression #needed model
from sklearn.neighbors import KNeighborsClassifier #
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
iris = load_iris() #loading database
data = iris.data
target = iris.target
```

```python
df = pd.DataFrame(data, columns=iris.feature_names)
df['target']= target
```

```python
sns.countplot(x='target',data=df)
plt.show()
```

```python
corr= df.corr()
sns.heatmap (corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
scalar=StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)
print(y_pred_logreg)
```

```python
def calculate_metrices (y_true, y_pred):
  accuracy = accuracy_score(y_true, y_pred)
  precision = precision_score(y_true, y_pred, average='weighted')
  f1 = f1_score(y_true, y_pred, average='weighted')
  recall=recall_score(y_true, y_pred, average='weighted')
  confusion= confusion_matrix(y_true, y_pred)
  return accuracy, precision,  f1, recall,confusion

accuracy_logreg, precision_logreg, recall_logreg, f1_logreg, confusion_logreg = calculate_metrices(y_test, y_pred_logreg)





```

```python
# Print Logistic Regression metrics
print("Logistic Regression metrics")
print("Accuracy:", accuracy_logreg)
print("Precision:", precision_logreg)
print("F1 score:", f1_logreg)
print("Recall:", recall_logreg)
print("Confusion Matrix:\n", confusion_logreg)

```
